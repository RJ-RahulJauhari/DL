```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('image.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Original Image")
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.subplot(1,2,2)
plt.imshow(gray_img, cmap='gray')
plt.title("Grayscale Image")
plt.show()
```

```python
hist1 = cv2.calcHist([gray_img], [0], None, [256], [0, 256])
```

```python
img_2 = cv2.equalizeHist(gray_img)
```

```python
hist2 = cv2.calcHist([img_2], [0], None, [256], [0, 256])
```

```python
plt.figure(figsize=(12, 8))
plt.subplot(2,2,1), plt.imshow(gray_img, cmap='gray'), plt.title('Original Grayscale Image')
plt.subplot(2,2,2), plt.plot(hist1), plt.title('Original Histogram')
plt.subplot(2,2,3), plt.imshow(img_2, cmap='gray'), plt.title('Equalized Grayscale Image')
plt.subplot(2,2,4), plt.plot(hist2), plt.title('Equalized Histogram')
plt.show()
```

```python
threshold_methods = [
    ('BINARY', cv2.THRESH_BINARY),
    ('BINARY_INV', cv2.THRESH_BINARY_INV),
    ('TRUNC', cv2.THRESH_TRUNC),
    ('TOZERO', cv2.THRESH_TOZERO),
    ('TOZERO_INV', cv2.THRESH_TOZERO_INV)
]
```

```python
plt.figure(figsize=(10, 8))
for i, (title, thresh_type) in enumerate(threshold_methods):
    _, th_img = cv2.threshold(gray_img, 200, 255, thresh_type)
    plt.subplot(2, 3, i+1)
    plt.imshow(th_img, cmap='gray')
    plt.title(title)
    plt.xticks([]), plt.yticks([])
plt.tight_layout()
plt.show()
```

```python
edges = cv2.Canny(gray_img, 100, 200)
plt.figure(figsize=(8, 6))
plt.imshow(edges, cmap='gray')
plt.title('Edges Detected')
plt.xticks([]), plt.yticks([])
plt.show()
```
